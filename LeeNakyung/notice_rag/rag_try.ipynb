```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader('./document.pdf')
docs = loader.load()
print(len(docs))

for doc in docs[:10]:
    print(doc.metadata)
    print(doc.page_content)
    print()
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# document.pdf chunking
splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=35,
    separators=['\n\n', '\n', '.', ' ', '']
)
docs = splitter.split_documents(docs)
docs
```

```python
print(len(docs)) # 17

for i, doc in enumerate(docs):
    print(f'{i}: page-{doc.metadata["page"]}')
    print(doc.page_content)
```

```python
# 임베딩 모델
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model='text-embedding-3-small')
```

```python
# vector store 생성
from langchain_chroma import Chroma

vector_store = Chroma.from_documents(
    documents=docs,
    embedding=embeddings
    )
vector_store
```

```python
# 검색 1: vector_store에 직접 질의
query = '청년일자리도약장려금사업이란 무엇인가요?'

# score: (코사인 거리) = 1 - 코사인유사도
# - 코사인 거리가 적을수록 유사한 데이터
retrievals = vector_store.similarity_search_with_score(query)

for i, (doc, score) in enumerate(retrievals):

    print(f'{i}: {score}')
    print(doc.page_content)
```

```python
# 검색 2: retriever 객체 사용
retriever = vector_store.as_retriever(
    search_type='similarity',
    search_kwargs={'k': 3}
)
retrievals = retriever.batch([query])

for i, doc in enumerate(retrievals[0]):
    print(f'{i}: {doc.page_content}')
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ('system',
        '당신은 공공 알림문·지원 제도 안내문이 어려울 때, 쉽고 빠르게 요약하고 질문에도 답해주는 문서 요약 + 질의응답 특화 챗봇입니다.'
        '질문하는 사람에게 최대한 호응해주며 context를 기반으로만 답변해 주세요.'),
    ('human', """
    제공된 context만을 참고하여, 다음 질문에 답변해 주세요.
    해당 context에서 확인할 수 없는 내용은 모른다고 답변해야 합니다.
    최종 응답에는 참조한 context에 대한 정보를 추가해 주세요.

    사용자 질문:
    {query}

    컨텍스트:
    {context}

    <<최종 응답 형식>>
    답변:

    참조 문서:
    - <<source>> (page: <<page>>): <<page_content>>
    """)
])
prompt
```

```python
prompt.invoke({'context': retriever, 'query': query})
```

```python
# 모델 생성
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model='gpt-4o-mini', temperature=0.1)
```

```python
# 체인 생성
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough

chain = {'context': retriever, 'query': RunnablePassthrough()} | prompt | llm | StrOutputParser()
print(chain.invoke(query))
```

```python
print(chain.invoke('근로조건은 무엇인가요?'))
```
